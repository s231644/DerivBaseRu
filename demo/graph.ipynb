```python
from collections import namedtuple, Counter
```

```python
Word = namedtuple('Word', ['token', 'lemma', 'pos', 'tags', 'dep', 'role'])
```

```python
import os
```

```python
c = Counter()
```

```python
sents = []

for doc in os.listdir('tagged_texts'):
    f = open(f'tagged_texts/{doc}')

    sent = []
    for line in f:
        if line.startswith('#'): 
            continue
        if len(line) < 3:
            sents.append(sent)
            sent = []
            continue
        l = line.split('\t')
        sent.append(Word(l[1], l[2], l[3], l[5], l[6], l[7]))
    if len(sents) >= 100:
        break
```

```python
sents[-1]
```

```python
vocabulary = set()
```

```python
for sent in sents:
    for token in sent:
        if token.pos in ['NOUN', 'ADJ', 'VERB', 'ADV']:
            vocabulary.add((token.lemma.lower(), token.pos.lower()))
```

```python
vocabulary
```

```python
vocab = list(vocabulary)
```

```python
print(len(vocab))
```

```python
arrows = []
for i in range(len(vocab)):
    source = vocab[i]
    print(source)
    if source[1] == 'verb':
        derived = set(evaluator.derive_node_ext(word_b=source[0], pos_b=source[1]))
    elif source[1] == 'adv':
        derived = set(evaluator.derive_node_ext(word_b=source[0], pos_b=source[1]))
    elif source[1] == 'noun':
        r = evaluator.tag_guesser.guess(word=source[0], pos_b='noun')
        if len(r):
            derived = set(evaluator.derive_wtags_node_ext(word_b=source[0], **r[0]))
        else:
            derived = set(evaluator.derive_node_ext(word_b=source[0], pos_b=source[1]))
    elif source[1] == 'adj':
        r = evaluator.tag_guesser.guess(word=source[0], pos_b='adj')
        if len(r):
            derived = set(evaluator.derive_wtags_node_ext(word_b=source[0], **r[0]))
        else:
            derived = set(evaluator.derive_node_ext(word_b=source[0], pos_b=source[1]))
    for der in derived:
        if der in vocabulary and der != source:
            arrows.append((source, der))
```
